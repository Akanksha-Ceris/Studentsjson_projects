Create an Iterator


to create an object/class as an iterator you have to implement the methods _iter_() and _next_() to your object.


```python
# create an iterator that returns numbers , starting with 1 and each sequence will increase by one (returning 1,2,3,4,5 etc)
```

```python
class MyNumbers:
  def _iter_(self):
    self.a = 1
    return self
  def _next_(self):
    x = self.a
    self.a +=1
    return x
myclass = MyNumbers()
myiter = iter(myclass)
print(next(myiter))

print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```

```python
#stop after 20 iterations:
class MyNumbers:
  def _iter_(self):
    self.a = 1
    return self
  def _next_(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration
myclass = MyNumbers()
myiter = iter(myclass)
for x in myiter:
  print(x)
```
